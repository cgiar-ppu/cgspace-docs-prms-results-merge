This file will generate the PDF contents file for the assessment of Golden IPSR levels given by Edwin. Using the attachment text, it will be uploaded to PPU-AI tool along with the developed prompt to verify the efficacy of prompt which will lead to further revisions (if needed)

```python
import os
import pandas as pd
```

### Reading the Combined PRMS and CGSpace Results

```python
prms_cgspace_results_dir = "prms_cgspace_merged"
prms_cgspace_results_file = os.path.join(prms_cgspace_results_dir, "prms_cgspace_merged_data.xlsx")
prms_cgspace_results = pd.read_excel(prms_cgspace_results_file, engine="openpyxl")
```

```python
prms_cgspace_results.head()
```

### Reading the Golden IPSR Assessment data

```python
golden_ipsr_input_dir = "test_input_data"
golden_ipsr_input_file = os.path.join(golden_ipsr_input_dir, "golden_ipsr_assessment.csv")
golden_ipsr_data = pd.read_csv(golden_ipsr_input_file)
golden_ipsr_data.head()
```

```python
golden_ipsr_data.info()
```

### Reading the Golden IPSR Assessment data

```python
golden_ipsr_data_attachments = pd.merge(golden_ipsr_data, prms_cgspace_results, how='left',left_on='Result Code', right_on='result_code', suffixes=('_ipsr', '_cgspace_prms')).reset_index(drop=True)
golden_ipsr_data_attachments
```

Excluding all the Sharepoint links because of unavailability of scraped data for now

```python
golden_ipsr_data_attachments = golden_ipsr_data_attachments[(golden_ipsr_data_attachments['attachment_text'].notnull()) & (golden_ipsr_data_attachments['attachment_part'].notnull())]
golden_ipsr_data_attachments
```

Extracting data only for Phase Year of 2023 since that is the source of Edwin Documents. Also get the latest version of the documents in Phase Year 2023

```python
def extract_required_phase_year_version_id(group_data):
    # Extract Phase Year 2023 and maximum Version ID. Then sort by Index and attachment_part
    group_data = group_data[group_data['phase_year'] == 2023]
    group_data = group_data[group_data['version_id'] == group_data['version_id'].max()]
    return group_data.sort_values(by=['Index', 'attachment_part'], ascending=[True, True])

golden_ipsr_filtered = golden_ipsr_data_attachments.groupby('Result Code').apply(extract_required_phase_year_version_id).reset_index(drop=True)
# Linking all the evidences under the same Result Code as different parts
golden_ipsr_filtered['Part'] = golden_ipsr_filtered.groupby('Result Code').cumcount() + 1
golden_ipsr_filtered['Content'] = golden_ipsr_filtered['attachment_text']
golden_ipsr_filtered
```

### Extracting the relevant columns to be passed to PPU-AI tool

```python
columns_extract = [
 'Result Code',
 'Short title',
 'valid_link',
 'Index',
 'Part',
 'Content']

golden_ipsr_extract = golden_ipsr_filtered[columns_extract]
golden_ipsr_extract.head()
```

```python
golden_ipsr_output_dir = "test_output_data"
os.makedirs(golden_ipsr_output_dir, exist_ok=True)
golden_ipsr_output_file = os.path.join(golden_ipsr_output_dir, "golden_ipsr_ppu_ai.xlsx")
golden_ipsr_extract.to_excel(golden_ipsr_output_file, index=False, engine="openpyxl")
```
